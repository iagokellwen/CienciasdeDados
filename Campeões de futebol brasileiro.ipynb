```python
import pandas as pd

from matplotlib import pyplot as plt

path = "C:/Users/autologon/campoes_brasileiro.csv"

#https://www.kaggle.com/datasets/ravels1991/campees-do-campeonato-brasileiro-de-futebol/

campeoes = pd.read_csv(path)

campeoes
```

```python
#Filtrando os clubes campeões
filtro = campeoes['Ano']>1986
```

```python
clubes = campeoes[filtro]
clubes[['Ano', 'Vencedor']]
```

```python
#Filtrando os clubes campeões
filtro = clubes['Ano']<1998
```

```python
clubes2 = clubes[filtro]
clubes2[['Ano', 'Vencedor']]
```

```python
#filtro = clubes2['Vencedor'] == 'Sport'
vencedores = clubes2.groupby(['Vencedor']).size()
vencedores
```

```python
vencedores.plot.barh()
plt.suptitle('Campeões de futebol brasileiro entre o ano de 1987 à 1997')
plt.ylabel('Vencedores')
plt.show()
```
