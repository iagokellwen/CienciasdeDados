```python
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams ['figure.figsize'] = 6,5
```

```python
x = range(1,11) 
y = [2, 10, 5, 4, 7, 2, 6, 6, 2, 10]
```

```python
# Plotar um grafico de linha
plt.plot(x,y, color='orange', ls='dashed', lw=2, marker="+", mew=5)
plt.xlabel('Rodadas')
plt.ylabel('Gols Marcados')
plt.title('Gols por rodadas marcados pelo Ceará em 2019')
```

```python
plt.barh(x,y, color='salmon')
#plt.savefig('rithely.png')
```
